```python
import pandas as pd
import matplotlib.pyplot as plt
team_data = pd.read_csv("/content/drive/MyDrive/OSUHackathon2025_TeamGameData.csv")

playoff_teams = team_data[team_data['WasCFPTeam'] == 1]
playoff_teams_unique = playoff_teams.drop_duplicates(subset=['GameId'])

playoff_teams_unique_2 = playoff_teams_unique[['Season', "GameId", "Team",
                                                "OppTeam", "Conference",
                                                "WasCFPTeam", "Score",
                                                "OppScore", "OffensiveSuccessRate_All",
                                                "OffensiveSuccessRate_Pass", "OffensiveSuccessRate_Run",
                                               "OffensiveBoomRate_All",
                                               "OffensiveBoomRate_Pass", "OffensiveBoomRate_Run",
                                               "OffensiveEPAPerPlay_Pass", "OffensiveEPAPerPlay_Run"]]
print(playoff_teams_unique_2)
```

```python
playoff_teams_unique_2
```

```python
# Playoff Teams per year gotten from: https://collegefootballplayoff.com/sports/2019/5/22/history

df_playoff_teams_2018 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2018) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2018

df_playoff_teams_2019 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2019) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2019

df_playoff_teams_2020 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2020) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2020

df_playoff_teams_2021 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2021) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2021

df_playoff_teams_2022 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2022) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2022

df_playoff_teams_2023 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2023) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2023

df_playoff_teams_2024 = playoff_teams_unique_2[(playoff_teams_unique_2['Season'] == 2024) &
                                                (playoff_teams_unique_2['WasCFPTeam'] == 1)]
df_playoff_teams_2024
```

```python
# EPA Per Pass Play Scatterplot by Year

mean_EPA_Pass_2019 = df_playoff_teams_2019['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2018 = df_playoff_teams_2018['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2021 = df_playoff_teams_2021['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2020 = df_playoff_teams_2020['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2022 = df_playoff_teams_2022['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2023 = df_playoff_teams_2023['OffensiveEPAPerPlay_Pass'].mean()
mean_EPA_Pass_2024 = df_playoff_teams_2024['OffensiveEPAPerPlay_Pass'].mean()

mean_EPA_Pass_years = [mean_EPA_Pass_2018, mean_EPA_Pass_2019, mean_EPA_Pass_2020, mean_EPA_Pass_2021, mean_EPA_Pass_2022, mean_EPA_Pass_2023, mean_EPA_Pass_2024]

plt.scatter([2018, 2019, 2020, 2021, 2022, 2023, 2024] , [mean_EPA_Pass_years])

print(mean_EPA_Pass_years)



# EPA Per Run Play ScatterPlot by Year

mean_EPA_Run_2019 = df_playoff_teams_2019['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2018 = df_playoff_teams_2018['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2020 = df_playoff_teams_2020['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2021 = df_playoff_teams_2021['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2022 = df_playoff_teams_2022['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2023 = df_playoff_teams_2023['OffensiveEPAPerPlay_Run'].mean()
mean_EPA_Run_2024 = df_playoff_teams_2024['OffensiveEPAPerPlay_Run'].mean()

mean_EPA_Run_years = [mean_EPA_Run_2018, mean_EPA_Run_2019, mean_EPA_Run_2020, mean_EPA_Run_2021, mean_EPA_Run_2022, mean_EPA_Run_2023, mean_EPA_Run_2024]

plt.scatter([2018, 2019, 2020, 2021, 2022, 2023, 2024] , [mean_EPA_Run_years])
plt.title("EPA Per Play Scatterplot by Year (Excluding 2020)")
plt.xlabel("Year")
plt.ylabel("EPA Per Play")

print(mean_EPA_Run_years)
```
